```python
import pandas as pd
import base64
import os
pd.set_option("display.max_colwidth", None)

# change to your file path
filename = os.path.join("..", "ebay_challenge", "desc_0.csv-002")
# filename = "/Users/chenyian261/Documents/ebay_challenge/desc_0.csv-002"
chunksize = 10**3

def decode_base64(encoded_string):
    decoded_bytes = base64.b64decode(encoded_string)
    return decoded_bytes.decode("utf-8")


with pd.read_csv(filename, chunksize=chunksize) as reader:
    for chunk in reader:
        print(chunk)

        chunk["ITEM_DESC"] = chunk["ITEM_DESC"].apply(decode_base64)

        with open('output.html', 'w') as file:
            html_text = chunk.loc[1]["ITEM_DESC"]
            file.write(html_text) 
        break
```
