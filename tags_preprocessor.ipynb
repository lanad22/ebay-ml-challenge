```python
# get all unique keys 
unique_keys = set()
for row in tags_df["ITEM_TAGS"]:
    for item in row:
        unique_keys.update(item.keys())
```

```python
print(len(unique_keys))
with open("tags_key.txt", "w") as f:
    f.write(",".join(list(unique_keys)))
unique_keys
```

```python
processed_tags_df = pd.DataFrame(columns=list(unique_keys))
```

```python
rows = []
# use values from the list of dictionaries
for index, row in tags_df.iterrows():
    row_data = {key: None for key in unique_keys}  
    for item in row["ITEM_TAGS"]:
        row_data.update(item)  
    rows.append(pd.DataFrame([row_data])) # Convert to DataFrame and add to list
# Create a new dataframe from the list of dictionaries
processed_tags_df = pd.DataFrame(rows)
```
